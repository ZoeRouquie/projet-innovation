```python
import pandas as pd

inde = pd.read_stata('C:/Users/zoero/OneDrive/Bureau/M2/S1/innovation/projet/Selection 2024-25/India-2014-Innovation-data.dta')
bangladesh = pd.read_stata('C:/Users/zoero/OneDrive/Bureau/M2/S1/innovation/projet/Selection 2024-25/Bangladesh-2013-Innovation-data.dta')
nepal = pd.read_stata('C:/Users/zoero/OneDrive/Bureau/M2/S1/innovation/projet/Selection 2024-25/Nepal-2013-Innovation-data.dta')
pakistan = pd.read_stata('C:/Users/zoero/OneDrive/Bureau/M2/S1/innovation/projet/Selection 2024-25/Pakistan-2013-Innovation-data-.dta')
nepal2 = pd.read_stata('C:/Users/zoero/OneDrive/Bureau/M2/S1/innovation/projet/Selection 2024-25/Nepal-2013-full-data-expanded.dta')
```

```python
inde.columns       = inde.columns.str.lower()
bangladesh.columns = bangladesh.columns.str.lower()
nepal.columns      = nepal.columns.str.lower()
pakistan.columns   = pakistan.columns.str.lower()
nepal2.columns     = nepal2.columns.str.lower()

# 3. Ajouter (ou créer) les colonnes a6a, a6b, b5 dans 'nepal' depuis 'nepal2'
colonnes_a_ajouter = ['a6a', 'a6b', 'b5']
for col in colonnes_a_ajouter:
    # Vérifier que la colonne existe dans nepal2
    if col in nepal2.columns:
        # Si la colonne n'existe pas dans nepal, on la crée
        if col not in nepal.columns:
            nepal[col] = nepal2[col]
        else:
            # Si besoin, on pourrait remplacer la colonne existante
            # mais selon votre description, ces colonnes n'existent pas dans nepal
            pass
```

```python
inde['country']       = 'IN'
bangladesh['country'] = 'BA'
nepal['country']      = 'NE'
pakistan['country']   = 'PA'
# (On n'a pas besoin d'ajouter 'country' à nepal2,
# puisqu'on ne concatène pas directement nepal2, on l'a juste utilisé pour récupérer des colonnes.)

# 5. Concaténer les bases en conservant uniquement les colonnes communes (join="inner")
merged_df = pd.concat([inde, bangladesh, nepal, pakistan], join="inner", ignore_index=True)

# 6. (Optionnel) Vérifier les colonnes non incluses
cols_inde       = set(inde.columns)
cols_bangladesh = set(bangladesh.columns)
cols_nepal      = set(nepal.columns)
cols_pakistan   = set(pakistan.columns)
cols_merged     = set(merged_df.columns)

non_included_columns = {
    'inde':       cols_inde       - cols_merged,
    'bangladesh': cols_bangladesh - cols_merged,
    'nepal':      cols_nepal      - cols_merged,
    'pakistan':   cols_pakistan   - cols_merged
}

print("Colonnes non incluses dans la fusion :")
for country, cols in non_included_columns.items():
    print(f"{country} : {cols}")
```

Traiter les auters colonnes

```python
import numpy as np

for col in merged_df.columns:
    if col in ["hb1", "hc4", "he1a", "he1b", "he1c","hd3a","hd3b","hd3c","hd4","hd7","hd8","hd5","he1d","he1e","he1f","he1g","he1h","he1i","he1j"]:
        # Remplacement des valeurs spécifiées
        merged_df[col] = merged_df[col].replace({
            "No": 0, 
            "Yes": 1
        })

        # Remplacer toutes les autres valeurs par NaN
        allowed_values = [0, 1]  # Liste des valeurs autorisées
        merged_df[col] = merged_df[col].where(merged_df[col].isin(allowed_values), np.nan)

```

Merger HE1: Marketing innovation

```python
# Liste des colonnes cibles
target_columns = ["he1a", "he1b", "he1c", "he1d","he1e","he1f","he1g","he1h","he1i","he1j"]

# Ajout d'une nouvelle colonne 'inno'
merged_df["he1-bis"] = None  # Initialise la colonne

for index, row in merged_df.iterrows():
    # Sélectionner uniquement les colonnes dans target_columns
    responses = row[[col for col in merged_df.columns if col in target_columns]]
    
    # Compter les occurrences des différentes réponses
    count_1 = (responses == 1).sum()
    count_0 = (responses == 0).sum()
    
    # Décider de la valeur de 'inno' pour cet individu
    if count_1 > 0:
        merged_df.at[index, "he1-bis"] = 1  # Si au moins un '1', inno = 1
    elif count_0 >= 6:  # Modifier cette condition selon vos besoins
        merged_df.at[index, "he1-bis"] = 0  # Si majorité de '0', inno = 0
    else:
        merged_df.at[index, "he1-bis"] = np.nan  # Sinon, inno = null

# Afficher un aperçu des résultats
print(merged_df[["he1-bis"]])
```

```python
# Liste des colonnes cibles
target_columns = ["he1-bis","hb1","hc4","hd-bis"]

# Ajout d'une nouvelle colonne 'inno'
merged_df["inno"]=None

for index, row in merged_df.iterrows():
    # Sélectionner uniquement les colonnes dans target_columns
    responses = row[[col for col in merged_df.columns if col in target_columns]]
    
    # Compter les occurrences des différentes réponses
    count_1 = (responses == 1).sum()
    count_0 = (responses == 0).sum()
    count_other = (responses == "other").sum()
    count_null = responses.isnull().sum()
    
    # Décider de la valeur de 'inno' pour cet individu
    if count_1 > 0:
        merged_df.at[index, "inno"] = 1  # Si au moins un '1', inno = 1
    elif count_0 >= 2:  
        merged_df.at[index, "inno"] = 0  # Si majorité de '0', inno = 0
    else:
        merged_df.at[index, "inno"] = np.nan  # Sinon, inno = null

# Afficher un aperçu des résultats
print(merged_df["inno"])
```
