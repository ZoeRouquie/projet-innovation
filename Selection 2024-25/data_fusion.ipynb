```python
import numpy as np

for col in merged_df.columns:
    if col in ["hb1", "hc4", "he1a", "he1b", "he1c", "he1d","he1e","he1f","he1g","he1h","he1i","he1j"]:
        # Remplacement des valeurs spécifiées
        merged_df[col] = merged_df[col].replace({
            "No": 0, 
            "Yes": 1, 
            "DONT KNOW": "other",
            "Dont know (Spontaneous)": "other",
            "DOESN\x92T APPLY": "other",
            "DON'T KNOW (SPONTANEOUS)": "other",
            "DON'T KNOW(SPONTANEOUS)": "other",
            "DON'T KNOW(SPONT.)": "other",
            "DON\x92T KNOW (SPONTANEOUS)": "other",
            "Does not apply (Spontaneous)": "other",
            "Doesn't Apply": "other",
            "DON\x92T KNOW": "other",
            "Dont Know": "other"
        })

        # Remplacer toutes les autres valeurs par NaN
        allowed_values = [0, 1, "other"]  # Liste des valeurs autorisées
        merged_df[col] = merged_df[col].where(merged_df[col].isin(allowed_values), np.nan)

```

Traitement de HE1

```python
# Liste des colonnes cibles
target_columns = ["he1a", "he1b", "he1c", "he1d","he1e","he1f","he1g","he1h","he1i","he1j"]

# Ajout d'une nouvelle colonne 'inno'
merged_df["he1-bis"] = None  # Initialise la colonne

for index, row in merged_df.iterrows():
    # Sélectionner uniquement les colonnes dans target_columns
    responses = row[[col for col in merged_df.columns if col in target_columns]]
    
    # Compter les occurrences des différentes réponses
    count_1 = (responses == 1).sum()
    count_0 = (responses == 0).sum()
    count_other = (responses == "other").sum()
    count_null = responses.isnull().sum()
    
    # Décider de la valeur de 'inno' pour cet individu
    if count_1 > 0:
        merged_df.at[index, "inno"] = 1  # Si au moins un '1', inno = 1
    elif count_0 >= 6:  # Modifier cette condition selon vos besoins
        merged_df.at[index, "inno"] = 0  # Si majorité de '0', inno = 0
    else:
        merged_df.at[index, "inno"] = np.nan  # Sinon, inno = null

# Afficher un aperçu des résultats
print(merged_df[["inno"]])
```

# Variables de contrôle

```python
merged_df["ha0"] = merged_df["ha0"].replace({
            "Services": 0, 
            "Service": 0,
            "Other Service":0, 
            "Manufacturing": 1,
            "Production":2
        
        })
```

```python

secteur_dummies = pd.get_dummies(merged_df["ha0"], prefix="secteur", drop_first=True)
secteur_dummies['secteur_1'].replace({
    True: 1,
    False:0
})

secteur_dummies['secteur_2'].replace({
    True: 1,
    False:0
})

secteur_dummies.rename(columns={
    "secteur_1": "Manufacturing",
    "secteur_2":"Production"
    }, inplace=True)

merged_df = pd.concat([merged_df, secteur_dummies], axis=1)
```

```python
new_data = pd.concat([merged_df['idstd'],merged_df['inno'],merged_df['Manufacturing'],merged_df['Production']],axis=1)
```
