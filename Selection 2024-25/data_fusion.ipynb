Traiter les auters colonnes

```python
import numpy as np

for col in merged_df.columns:
    if col in ["hb1", "hc4", "he1a", "he1b", "he1c","hd3a","hd3b","hd3c","hd4","hd7","hd8","hd5","he1d","he1e","he1f","he1g","he1h","he1i","he1j"]:
        # Remplacement des valeurs spécifiées
        merged_df[col] = merged_df[col].replace({
            "No": 0, 
            "Yes": 1
        })

        # Remplacer toutes les autres valeurs par NaN
        allowed_values = [0, 1]  # Liste des valeurs autorisées
        merged_df[col] = merged_df[col].where(merged_df[col].isin(allowed_values), np.nan)

```

Merger HE1: Marketing innovation

```python
# Liste des colonnes cibles
target_columns = ["he1a", "he1b", "he1c", "he1d","he1e","he1f","he1g","he1h","he1i","he1j"]

# Ajout d'une nouvelle colonne 'inno'
merged_df["he1-bis"] = None  # Initialise la colonne

for index, row in merged_df.iterrows():
    # Sélectionner uniquement les colonnes dans target_columns
    responses = row[[col for col in merged_df.columns if col in target_columns]]
    
    # Compter les occurrences des différentes réponses
    count_1 = (responses == 1).sum()
    count_0 = (responses == 0).sum()
    
    # Décider de la valeur de 'inno' pour cet individu
    if count_1 > 0:
        merged_df.at[index, "he1-bis"] = 1  # Si au moins un '1', inno = 1
    elif count_0 >= 6:  # Modifier cette condition selon vos besoins
        merged_df.at[index, "he1-bis"] = 0  # Si majorité de '0', inno = 0
    else:
        merged_df.at[index, "he1-bis"] = np.nan  # Sinon, inno = null

# Afficher un aperçu des résultats
print(merged_df[["he1-bis"]])
```
